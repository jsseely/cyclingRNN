```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_eigs(A_mat):
  """
    Docstring
  """
  w, _ = np.linalg.eig(A_mat)
  re_w = np.real(w)
  im_w = np.imag(w)
  plt.figure(figsize=(5, 5))
  plt.plot(re_w, im_w, 'o')
  theta = np.linspace(0, 2*np.pi, num=50)
  x_cir = np.cos(theta)
  y_cir = np.sin(theta)
  plt.plot(x_cir, y_cir, 'k', linewidth=0.5, alpha=0.5)
  plt.plot([-100, 100], [0, 0], 'k', linewidth=0.5, alpha=0.5)
  plt.plot([0, 0], [-100, 100], 'k', linewidth=0.5, alpha=0.5)
  plt.xlim([-1.5, 1.5])
  plt.ylim([-1.5, 1.5])
```

```python
n = 50
A = np.random.randn(n,n)/np.sqrt(n)
plot_eigs(A)
```

```python
A = np.random.multivariate_normal(n**2*[0], np.eye(n**2)/n)
```

```python
A = A.reshape((n,n))
```

```python
plot_eigs(A)
```
